```python
def pandas2tsv(df,path):
    base_path = "D:/Users/ldavid/Documents/Stock Combustible"
    sep = ','
    index = False
    header = True
    encoding = 'utf-8-sig'
    df.to_csv(base_path+path,
              index=index,
              sep=sep,
              header=header,
              encoding=encoding
              )

    
```
